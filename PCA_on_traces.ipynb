```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
fname = 'events.csv'
events = np.loadtxt(fname, delimiter=',')
event = events[:500, 2] # look at one trace
```

### Create a new data variable X by scanning the trace

```python
X = []
window_size = 10
for i in range(len(event) - window_size):
    x = []
    for j in range(window_size):
        x.append(event[i + j])
    X.append(x)
X = np.asarray(X)
```

### Perform PCA on scanned trace

```python
n_components = 5
pca = PCA(n_components = n_components)
scores = pca.fit_transform(X)
variance = pca.explained_variance_ratio_

fig = plt.figure(figsize = (12, 15))
for i in range(n_components):
    fig.add_subplot(n_components, 1, 1 + i)
    plt.plot(scores[:, i])
    plt.title('PCA: ' + str(i + 1) + ', Explained variance: ' + str(variance[i] * 100)[:4] + '%')
plt.tight_layout()
```
